Use Numpy library


```python
import numpy as np
```

```python
#1
a = np.random.randint(1,101,20)
b = np.random.randint(1,101,20)
print(a)
print(b)
```

```python
#2
a = a.reshape(5,4)
b = b.reshape(5,4)
print(a)
print(b)
```

```python
#3
a = a * 3
b = b * 3
print(a)
print(b)
```

```python
#4
b = b.reshape(4,5)
c = np.dot(a,b)
print(c)
```

```python
#5
print(c[:,1:3])
print(c[:])
```

```python
#6
elem = np.max(c)
index = np.unravel_index(np.argmax(c), c.shape)
print(elem)
print(index)
```

```python
#7
c = c.flatten()
print(c)
```

Use Matplotlib Library

```python
import matplotlib.pyplot as plt
```

```python
#1
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]

plt.figure(figsize=(9,7))
plt.scatter(temperature_dhaka, humidity_dhaka, marker= "*", color = "r")
plt.xlabel("temperature")
plt.ylabel("humidity")
plt.show()
```

```python
#2
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

plt.figure(figsize=(12,8))
plt.plot(study_hours, marks, marker = "*", linestyle= "--", color= "black")
plt.xlabel("study_hours")
plt.ylabel("marks")
plt.show()

```

```python
#3
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]

plt.barh(subjects, marks, color= "orange")
plt.ylabel("subjects")
plt.xlabel("marks")
plt.show()
```

```python
#4
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]

colors= ["red", "blue", "green", "pink", "grey"]
plt.bar(subjects, marks, color= colors)
plt.xlabel("subjects")
plt.ylabel("marks")
plt.show()
```
